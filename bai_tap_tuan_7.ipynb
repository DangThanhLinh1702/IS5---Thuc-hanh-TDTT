```python
#W7A1
# Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần.
# Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def
# binary_search(arr, left, right, target)

def binary_search(arr, left, right, target):
    left = 0
    right = len(arr) - 1

    while left <= right:    
        mid = (left + right)//2

        if arr[mid] == target:
            return mid
        elif arr[mid] > target: #target o ben trai
            right = mid - 1
        elif arr[mid] < target: #target o ben phai
            left = mid + 1
    else:
        return -1
    
n = list(map(int, input().split()))
target = int(input())
print(binary_search(n,0,len(n)-1,target))

```

```python
#W7A2
# Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần
# trong danh sách. def count_occurrences(arr, x) - > int:

def count_occurrences(arr,x) -> int:
    cnt = 0
    for i in arr:
        if i == x:
            cnt += 1
    return cnt

n = list(map(int, input().split()))
x = int(input())
print(count_occurrences(n,x))
```

```python
# Sắp xếp chọn (Selection Sort)
# Ngược với bài trên, lần này ta cho phần tử nhỏ nhất dần lên đầu

def selection_sort(lst):
    n = len(lst)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1,n):
            if lst[j] < lst[min_idx]:
                min_idx = j
        lst[i], lst[min_idx] = lst[min_idx], lst[i]
    return lst

n = list(map(int, input().split()))
print(selection_sort(n))
```

```python
# W7A4
# Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có
# nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.

def appear_time(num,lst):
    cnt = 0
    for i in lst:
        if i == num:
            cnt += 1
    return cnt
def most_appear(lst):
    max_appear = 0
    min_num = 0 
    for num in lst:
        if appear_time(num,lst) > max_appear or (appear_time(num,lst) == max_appear and num < min_num):
            max_appear = appear_time(num,lst)
            min_num = num
    return f'So xuat hien nhieu nhat: {min_num}, so lan xuat hien: {max_appear}'
n = list(map(int, input().split()))
print(most_appear(n))
```

```python
# W7A6
# Bạn được cung cấp một danh sách các danh sách lồng nhau, vd [[1, 2, 3], [4,5], [6, 7, 8, 9]],
# hãy làm phẳng danh sách

def flat_list(arr):
    result = []
    for lst in arr:
        for num in lst:
            result.append(num)
    return result


print(flat_list([[1, 2, 3], [4, 5], [6, 7, 8, 9]]))
```

```python
# W7A7
# Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng
# nghiêm ngặt dài nhất (không nhất thiết phải liền kề).

#xét từng dần từng thằng xem có bao nhiêu thằng bé hơn thằng đấy
line = list(input().split()) #[10, 9, 2, 5, 3, 7, 101, 18]

def len_strictly_increase(arr):
    n = len(arr) 
    stor = [1] * n
    for i in range(n):
        for j in range(i):
            if arr[j] < arr[i]:
                stor[i] = max(stor[i], stor[j]+1)
    return max(stor)

print(len_strictly_increase(line))
```

```python
# W7A8
# Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...]
# và một danh sách các truy vấn queries = [q1, q2, ...] Với mỗi truy vấn q, hãy tìm độ dài
# nhỏ nhất của khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri. Nếu không tồn tại khoảng thời 
# gian nào như vậy, hãy trả về -1 cho truy vấn đó.
# solve in < O(m.n), hint: binary search in the queries list

def check(arr,lst):
    res = []
    for i in range (len(lst)):
        temp = []
        for j in range (len(arr)):
            if lst[i] >= arr[j][0] and lst[i] <= arr[j][1]:
                temp.append(arr[j][1] - arr[j][0] + 1)
        if len(temp) == 0:
            res.append(-1)
        else :
            res.append(min(temp))
    return res
arr = [[1,3],[5,7],[9,12]]
lst = [2,4,5,8,10]
print(check(arr,lst))
        
```

```python
# W7A10
# Cho một danh sách các số
# nguyên nums và một số
# nguyên m, hãy tìm tổng
# mảng con lớn nhất theo
# modulo m.
# Tức là, với bất kỳ mảng
# con nums[i:j] nào, hãy tính
# sum(nums[i:j]) % m và trả
# về kết quả lớn nhất có thể.
# Gợi ý:
# Duy trì tổng tiền tố (prefix
# sum) mod m và sử dụng
# tìm kiếm nhị phân để tìm
# tổng tiền tố nhỏ nhất chỉ
# lớn hơn tổng hiện tại một
# chút (để tối đa hóa hiệu số
# bao quanh).

nums = [7, 11, 14, 2, 8, 4] 
m = 12
def largest_subarray_sum(nums,m):
    n = len(nums)
    res = []
    for i in range(n):
        for j in range(n):
            res.append(sum(nums[j:i]) % m)
    return max(res)

print(largest_subarray_sum(nums,m))
```

```python
# W7A13
# Dãy hài hòa là một dãy
# thỏa mãn điều kiện giá trị
# lớn nhất và giá trị nhỏ nhất
# của dãy khác nhau đúng
# bằng \( 1 \).
# Viết hàm trả về độ dài của
# dãy hài hòa con dài nhất có
# thể có trong mảng giá trị
# nguyên \( arr \) có \( n \)
# phần tử
# Ví dụ:
# Input: 1 3 2 2 5 2 3 7
# Output: 5 - Dãy hài hòa con
# dài nhất là [3,2,2,2,3]

def harmony_sequence_len(seq):
    cnt = {}
    for num in seq:
        cnt[num] = cnt.get(num,0)+1
    
    max_len = 0
    for s in cnt:
        if s+1 in cnt:
            current_len = cnt[s] + cnt[s+1]
            max_len = max(max_len, current_len)
    return max_len

nums = list(map(int, input().split()))
print(harmony_sequence_len(nums))
```

```python
# W7A14
#  Hãy viết một chương
# trình tìm kiếm tất cả các số
# 7 có trong một dãy số.
# Đầu vào: số nguyên n trên
# một dòng và n số nguyên ở
# dòng kế tiếp.
# Đầu ra:
# Vị trí của các số \( 7 \)
# trong dãy cách nhau bởi
# dấu cách, vị trí lớn hơn
# đứng trước và bắt đầu tính
# từ \(0\).
# Nếu không tìm được số \( 7
# \) nào in ra "Not found" 

def finding_seven(line):
    res = []
    for idx, i in enumerate(line):
        if i == 7:
            res.append(idx)
    if not res:
        return"Not found"
    return res

n = int(input())
while True:
    line = list(map(int, input().split()))
    if len(line) == n:
        break
print(finding_seven(line))
```

```python
# W7A15
#  Hãy giúp Nemo tìm tên
# của các nhân vật này, biết
# rằng tên trong danh sách
# khách mời được sắp xếp
# theo vị trí ngồi.
# Đầu vào: số nguyên n - số
# lượng khách mời trong
# danh sách trên một dòng, n
# dòng tiếp theo là tên của
# các vị khách mời.
# Đầu ra: Tên của hai người
# ngồi cạnh Nemo theo định
# dạng "<name1> and
# <name2>" (name1 là tên
# của người đứng trước và
# name2 là tên của người đứng sau Nemo trong danh
# sách). 

def finding_nemo(names):
    for i in range(len(names)):
        if names[i] == "Nemo":
            if i == len(names) - 1:
                return f'{names[i-1]} and {names[0]}'
            else:
                return f'{names[i-1]} and {names[i+1]}'

n = int(input())
names = []
for _ in range(n):
    name = input()
    names.append(name)
print(finding_nemo(names))
```
