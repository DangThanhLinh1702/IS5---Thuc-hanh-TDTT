```python
def uoc_nguyen_to_max(n):
    uoc_max=0
    i=2
    while i*i <= n: #Xét i từ 2 đến căn n
        while n%i==0:
            n//=i #n = phần nguyên của n/i
        i+=1 #nếu n chia hết i trước thì tăng i lên 1
    #lý do viết i+=1 ra ngoài while: while bên trong sẽ chia hết tất cả các số lặp lại (VD: 2x2x2x3), sau đó mới tăng i lên 1 đvi
    if n>1:
        uoc_max=n
    return uoc_max
    

n=int(input())
ket_qua = uoc_nguyen_to_max(n)
print(ket_qua)

        
```

```python
n=int(input())
result = []
for i in range(1,n+1):
    if i*i <= n:
        scp = i*i
        chu_so= str(scp)
        if len(set(chu_so)) == len(chu_so): #set(s) là tập hợp các ký tự khác nhau trong chuỗi s.
            result.append(chu_so)

if result:
    print(" ".join(result)) #" ".join(results) nối các phần tử trong danh sách results thành một chuỗi, ngăn cách nhau bằng một dấu cách " "
else: 
    print("no number")
```

```python
#LongestCollatzUpToN

def collatz_length(i):
    length = 1
    while i != 1: 
        if i%2==0:
            i = i/2
        else:
            i = 3*i + 1
        length += 1
    return length


n = int(input())
best_i =1
best_len = collatz_length(1)

for i in range (1,n+1):
    L = collatz_length(i)
    if L > best_len or (L == best_len and i < best_i): #đề bảo có cùng độ dài thì chọn số nhỏ nhất
        best_len = L
        best_i = i

print (best_i, best_len)


```

```python
def reverse(n):
    return int(str(n)[::-1])
def is_palindrome(n):
    s=str(n)
    return s == s[::-1]
n=int(input())
while not is_palindrome(n):
    step=0
    n = n + reverse(n)
    step+= 1

print(step, n)
```

```python

def tong_uoc(x):
    s=0
    for i in range(1,int(x/2)+1):
        if x%i==0:
            s+=i
    return s

n=int(input())
m=int(input())
if tong_uoc(n) == m and tong_uoc(m) == n:
    print("True")
else:
    print("False")
```
