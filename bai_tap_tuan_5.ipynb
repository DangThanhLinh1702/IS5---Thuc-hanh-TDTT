```python
#W5A3
n = int(input())
def prime(n):
    if n < 2:
        return False
    for i in range (2, int(n**0.5) + 1):
        if n % i == 0:
            return False     
    return True
print(prime(n))

```

```python
#W5A5
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 
#bài này dùng indices thì ok
k = int(input())
lst = list(input())

def vi_tri_so_k(n):
    for idx, num in enumerate(lst):
        if int(num) == n:
            return idx+1
    else:
        return -1

vi_tri_so_k(k)


```

```python
#W5A6
#Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 
def giai_thua(n):
    gt = 1
    for i in range(1,n+1):
        gt*=i
    return gt

n=int(input())
print(giai_thua(n))
```

```python
#W5A7 
#Viết hàm giả lập một máy tính bỏ túi. 
def caculator(n1,n,n2):
    if n == "+":
        return round((n1+n2),2)
    if n == "-":
        return round((n1-n2),2)
    if n == "*" or "x":
        return round((n1*n2),2)
    if n == ":" or "/":
        return round((n1/n2),2)

num1= int(input("so thu nhat: "))
dau = input("nhap phep tinh")
num2= int(input("so thu hai: "))
print(caculator(num1,dau,num2))
```

```python
#W5A8
#tính khoảng cách Hamming
def binary(n):
    if n==0:
        return '0'
    nhi_phan=''
    while n!=0:
        nhi_phan = str(n%2) + nhi_phan 
        n = n//2
    return nhi_phan

n1= int(input())
n2= int(input())
print(binary(n1), "\n" ,binary(n2))

print(abs(len(binary(n1))-len(binary(n2))))

    
```

```python
#W5A9
#Viết hàm trả về tổng các chữ số của số nguyên dương n
def tong_chu_so(n):
    tong=0
    while n!= 0:
        tong= tong+n%10
        n=n//10
    return tong

n= int(input())
print(tong_chu_so(n))
```

```python
#W5A10
# hai từ đẳng cấu

                  #abca  và  zmnz
def cong_thuc(n):
    kgian_mau={}
    ket_qua = []
    for idx, char in enumerate(n):
        if char not in kgian_mau:
            kgian_mau[char] = idx       #kgian mẫu đã nhận 'a'==0 nên về sau khi gặp a sẽ cho ra số 0
        ket_qua.append(kgian_mau[char]) # nếu chữ lặp lại thì ko chạy if, result append luôn chữ đã gán trước đó
    return ket_qua

def dang_cau(n1,n2):
    if cong_thuc(n1) ==  cong_thuc(n2):
        print('True')
    else:
        print('False')

n1 = input()
n2 = input()
dang_cau(n1,n2)
            
            
    
```
